```python
import numpy as np
import numpy.matlib
import math
import timeit
```

```python
d=np.loadtxt('originaldata.txt')
t=d[0:140,:]
c1=np.zeros((1,8))
c2=np.zeros((1,8))
for i in range(140):
    if t[i,0]==0:
        c1=np.row_stack((c1,t[i,:]))
    else:
        c2=np.row_stack((c2,t[i,:]))
c1=c1[1:,1:7] 
c2=c2[1:,1:7]
nc1=len(c1)
nc2=len(c2)
t1=t[0:140,0]
t1=np.mat(t1)
t1=t1.T
t2=t[0:140,1:7]
```

```python
m1=np.mean(c1,axis=0)
m2=np.mean(c2,axis=0)
m1=np.mat(m1)
m2=np.mat(m2)
m11=np.zeros((1,6))
m22=np.zeros((1,6))
for i in range(len(c1)):
    m11=np.row_stack((m11,m1))
for i in range(len(c2)):
    m22=np.row_stack((m22,m2))
m11=m11[1:,0:6]
m22=m22[1:,0:6]
v1=(((c1-m11).T)@(c1-m11))/len(c1)
v2=((c2-m22).T)@(c2-m22)/len(c2)
p1=len(c1)/140
p2=len(c2)/140
```

```python
start = timeit.default_timer()
v=v1+v2
w=np.linalg.inv(v)@((m2-m1).T)# 得到列向量
g=t2@w #要使g为列向量（每个螃蟹的6个数据得一个常数）只能这样乘
print(g)
for i in range(len(g)):
    if g[i,0]>-7:
        g[i,0]=1
    else:
        g[i,0]=0 
stop = timeit.default_timer()
n1=0
n2=0
n3=0
n4=0
for i in range(len(g)):
    if g[i,0]==t1[i,0] and g[i,0]==0:
        n1=n1+1
    elif g[i,0]==t1[i,0] and g[i,0]==1:
        n2=n2+1
    elif g[i,0]==0:
        n3=n3+1
    elif g[i,0]==1:
        n4=n4+1

print('Training dataset')
print('         Classify 0   Classify 1')
print('Class 0   ',n1,'          ',n3)
print('Class 1   ',n4,'          ',n2)
print('Time',stop-start)
```
